```
from hashlib import sha256
from google.colab import files
import os 
import re
import tensorflow as tf
import numpy as np
from google.colab import files
```

```
def ascitobin(n):
    l=[]
    for i in n:        
        if i == 'a':
            l.append('01100001')
        elif i == 'b':
            l.append('01100010')
        elif i == 'c':
            l.append('01100011')
        elif i == 'd':
            l.append('01100100')
        elif i == 'e':
            l.append('01100101')
        elif i == 'f':
            l.append('01100110')
        elif i == 'g':
            l.append('01100111')
        elif i == 'h':
            l.append('01101000')
        elif i == 'i':
            l.append('01101001')
        elif i == 'j':
            l.append('01101010')
        elif i == 'k':
            l.append('01101011')
        elif i == 'l':
            l.append('01101100')
        elif i == 'm':
            l.append('01101101')
        elif i == 'n':
            l.append('01101110')
        elif i == 'o':
            l.append('01101111')
        elif i == 'p':
            l.append('01110000')
        elif i == 'q':
            l.append('01110001')
        elif i == 'r':
            l.append('01110010')
        elif i == 's':
            l.append('01110011')
        elif i == 't':
            l.append('01110100')
        elif i == 'u':
            l.append('01110101')
        elif i == 'v':
            l.append('01110110')
        elif i == 'w':
            l.append('01110111')
        elif i == 'x':
            l.append('01111000')
        elif i == 'y':
            l.append('01111001')
        elif i == 'z':
            l.append('01111010')
        
        else:
            s("\n ASCII error" +
                    str(n[i]), end = '')
    return(l)
```

```
def hash(text):
  sent = (text)
  hex=(sha256(sent.encode('utf-8')).hexdigest())
  return(hex)
```

```
def HexToBin(hexdec):
    l=[]
    for i in hexdec:        
        if i == '0':
            l.append('0'+'0'+'0'+'0')
        elif i == '1':
            l.append('0'+'0'+'0'+'1')
        elif i == '2':
            l.append('0'+'0'+'1'+'0')
        elif i == '3':
            l.append('0'+'0'+'1'+'1')
        elif i == '4':
            l.append('0'+'1'+'0'+'0')
        elif i == '5':
            l.append('0'+'1'+'0'+'1')
        elif i == '6':
            l.append('0'+'1'+'1'+'0')
        elif i == '7':
            l.append('0'+'1'+'1'+'1')
        elif i == '8':
            l.append('1'+'0'+'0'+'0')
        elif i == '9':
            l.append('1'+'0'+'0'+'1')
        elif i == 'A' or i == 'a':
            l.append('1'+'0'+'1'+'0')
        elif i == 'B' or i == 'b':
            l.append('1'+'0'+'1'+'1')
        elif i == 'C' or i == 'c':
            l.append('1'+'1'+'0'+'0')
        elif i == 'D' or i == 'd':
            l.append('1'+'1'+'0'+'1')
        elif i == 'E' or i == 'e':
            l.append('1'+'1'+'1'+'0')
        elif i == 'F' or i == 'f':
            l.append('1'+'1'+'1'+'1')
        else:
            s("\nInvalid hexadecimal digit " +
                    str(hexdec[i]), end = '')
    return(l) 
```

```
def shapre(word):
    wordbinlist=ascitobin(word)
    wordbin=""
    for j in wordbinlist:
      wordbin=wordbin+j
    c=0
    wordarr = ['0' for _ in range(256)]
    wordarr[255]=len(word)
    wordarr[len(wordbin)]='1'
    c=0
    for i in wordbin:
      wordarr[c]=i
      c=c+1
    return wordarr
```

```
uploaded = files.upload()
for fn in uploaded.keys():
    print('User uploaded file "{name}" with length {length} bytes'.format(
      name=fn, length=len(uploaded[fn])))

#upload the words from dataset from Kaggle

fpath = os.getcwd(); fpath
corpus = open(fpath + '/words.txt',  'rb').read().lower().decode(encoding='utf-8')
corpus = re.sub('\n', ' ', corpus)
corpus = re.sub('\r', ' ', corpus)
words = corpus.split(" ")
```

```
a=[]
wordlist=[]
c=0
s=0
for i in words:
  try:
    a.append(shapre(i))
    wordlist.append(i)
  except:
    s=s+1
  c=c+1
print('The number of words left out were={words}'.format(words=s))
print('The total number of words used for machine learning are={l}'.format(l=len(wordlist)))
```

```
#1st tensor (preprocessed words into binary before hash function)
y=[]
for i in a:
  temp=np.array(i)
  ytensor=tf.convert_to_tensor(temp, dtype=tf.float32)
  y.append(ytensor)

print(wordlist[0])
print(y[0])
```

```
#2nd tensor (sha256 hashed words in binary)
bin=[]
for i in wordlist:
  temp=i
  hashit=hash(temp)
  tobin=HexToBin(hashit)
  tempbin=""
  for i in tobin:
    tempbin=tempbin+i
  bin.append(tempbin)
b=[]
temp=[]
for i in bin:
  for j in i:
    if (j=='1'):
      temp.append(1)
    else:
      temp.append(0)
  b.append(temp)
  temp=[]
x=[]
for j in b:
  temp=np.array(j)
  xtensor=tf.convert_to_tensor(temp, dtype=tf.float32)
  x.append(xtensor)
print(wordlist[0])
print(x[0])
```

```
print(x[99])
print(y[99])
x=np.array(x)
y=np.array(y)
print(x.shape)
print(y.shape)
```

```
#the simplest tf model
model = tf.keras.Sequential([tf.keras.layers.Dense(256, activation='sigmoid')])
```

```
model.compile(loss='mean_squared_error', optimizer='sgd', metrics=['accuracy'])
```

```
history=model.fit(x, y, epochs=10)
```

```
acc = history.history['accuracy']
loss = history.history['loss']
import matplotlib.pyplot as plt

epochs = range(1, len(acc) + 1)

plt.plot(epochs, acc, 'b', label='Training acc')
plt.title('Training accuracy')
plt.legend()

plt.figure()

plt.plot(epochs, loss, 'b', label='Training loss')
plt.title('Training loss')
plt.legend()

plt.show()
```
